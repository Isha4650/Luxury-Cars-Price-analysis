```python
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
print("Before:", df.shape)
df.drop_duplicates(inplace=True)
print("After:",df.shape)
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned = df_cleaned.dropna(subset=["Engine HP", "Engine Cylinders", "Number of Doors"])
df
```

```python
df_cleaned["Market Category"] = df_cleaned["Market Category"].fillna("Unknown") 
df
```

```python
df.drop(['Market Category', 'Model',],axis=1,inplace=True)
df
```

```python
avg_msrp_by_make = df.groupby("Make")["MSRP"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
avg_msrp_by_make.plot(kind='bar', color='steelblue')
plt.title("Top 10 Car Brands by Average MSRP")
plt.ylabel("Average Price")
plt.xlabel("Car Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_cleaned, x="Engine HP", y="MSRP", hue="Vehicle Size", alpha=0.6)
plt.title("Engine Horsepower vs Price by Vehicle Size")
plt.xlabel("Engine Horsepower (HP)")
plt.ylabel("MSRP")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_cleaned, x="Vehicle Size", y="city mpg")
plt.title("City Mileage Distribution by Vehicle Size")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df_cleaned, x="Vehicle Size", y="highway MPG")
plt.title("Highway Mileage Distribution by Vehicle Size")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned["MSRP"], bins=40, kde=True)
plt.title("Distribution of Car Prices (MSRP)")
plt.xlabel("MSRP ($)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_cleaned.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
sns.pairplot(df[['Engine HP', 'Engine Cylinders', 'city mpg', 'highway MPG', 'MSRP']].dropna(), diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Car Features', y=1.02)
plt.show()
```
